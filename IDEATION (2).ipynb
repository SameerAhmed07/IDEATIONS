```python
#SOLAR ENERGY PREDICTION :
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

years = list(range(2010, 2023))
solar_energy_produced = [1000, 1200, 1400, 1800, 2200, 2500, 2600, 2800, 3100, 3400, 3700, 4000, 4200]

data = pd.DataFrame({'YEAR': years, 'SOLAR ENERGY PRODUCED': solar_energy_produced})

data.to_csv('solar_energy_data.csv', index=False)

data = pd.read_csv('solar_energy_data.csv')

X = data[['YEAR']]
y = data['SOLAR ENERGY PRODUCED']

model = LinearRegression()
model.fit(X, y)

future_years = list(range(2023, 2031)) 
future_predictions = model.predict(pd.DataFrame({'YEAR': future_years}))

plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Historical Data')
plt.plot(future_years, future_predictions, color='red', marker='o', linestyle='dashed', label='Predicted')
plt.title('Future Solar Energy Production Prediction')
plt.xlabel('Year')
plt.ylabel('Solar Energy Produced')
plt.legend()
plt.show()

for year, prediction in zip(future_years, future_predictions):
    print(f"Year: {year}, Predicted Solar Energy: {prediction}")
```

```python
#WIND ENERGY PREDICTION
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'AVG WIND SPEED': [4.5, 5.0, 5.5, 6.0, 6.5, 7.0, 7.5, 8.0, 8.5, 9.0, 9.5, 10.0, 10.5],
    'WIND ENERGY PRODUCED': [1000, 1200, 1400, 1800, 2200, 2500, 2600, 2800, 3100, 3400, 3700, 4000, 4200]
})
data.to_csv('wind_energy_data.csv', index=False)
data = pd.read_csv('wind_energy_data.csv')

X = data[['AVG WIND SPEED']]
y = data['WIND ENERGY PRODUCED']
model = LinearRegression()
model.fit(X, y)
future_wind_speeds = [11.0, 11.5, 12.0, 12.5, 13.0]
future_predictions = model.predict(pd.DataFrame({'AVG WIND SPEED': future_wind_speeds}))
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Historical Data')
plt.plot(future_wind_speeds, future_predictions, color='red', marker='o', linestyle='dashed', label='Predicted')
plt.title('Future Wind Energy Production Prediction')
plt.xlabel('Average Wind Speed')
plt.ylabel('Wind Energy Produced')
plt.legend()
plt.show()
for wind_speed, prediction in zip(future_wind_speeds, future_predictions):
    print(f"Avg Wind Speed: {wind_speed}, Predicted Wind Energy: {prediction}")
```
